### Question 1

```python
x = 1

def function1(x):
    return -2020*(x*x*x) + 2020*(x*x) - 4040

def derivative(x):
    return -6060*x^2 + 4040*x



value1 = function1(x)  #function output
value2 = derivative(x)  #derivative function output
value3 = value1/value2    #the gradient vector

threshold =  -1 #0.00000001

    
def raphson(output, derivative, x):
    
    gradient = output/derivative
    return (x - gradient)

if value3 > threshold:
    x = raphson(value1, value2, x)
    print("Root or the value close to the root", x)
    
```
